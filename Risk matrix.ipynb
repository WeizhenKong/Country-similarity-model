```python
import pandas as pd
import numpy as np
import os
import unicodedata
import calendar
import datetime
```

```python
file = pd.ExcelFile('KMU_risk_analysis_2020_01_09.xlsx')
df = pd.read_excel(file,'records')
```

```python
df
```

```python
for n in range(len(df['country'])):
    if str(df['country'][n]) != 'nan':
        continue
    else:
        df['country'][n] = df['country'][n-1]
```

```python
df
```

```python
def caldistance(vec1,vec2):
    dist = np.sqrt(np.sum(np.square(vec1 - vec2)))
    dist = float("{0:.2f}".format(dist))
    return dist
```

```python
def calculation2(n1,n2):
    v1 = np.array(df.iloc[n1][2:])
    v2 = np.array(df.iloc[n2][2:])
    grade = caldistance(v1,v2)
    return grade
```

```python
total = []
for i in range(0,len(df)):
    grade=[]
    for n in range(0,len(df)):
        grade.append(calculation2(i,n))
    total.append(grade)
```

```python
matrix = pd.DataFrame(total)
matrix['country'] = df['country']
matrix['year'] = df['year']
matrix
```
